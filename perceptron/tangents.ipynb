```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

# Define the function and its derivative
x_sym = sym.Symbol('x')
# func_sym = sym.sin(x_sym) ** 2  # Change the function here
# func_sym = x_sym ** 2
alpha = 0.01
func_sym = alpha * x_sym  # Change the function here
derivative_sym = diff(func_sym, x_sym)

# Display the function and its derivative in LaTeX format
display(Eq(sym.symbols('f(x)'), func_sym))
display(Eq(sym.symbols("f'(x)"), derivative_sym))

# Convert the symbolic derivative to a Python function
func = sym.lambdify(x_sym, func_sym)
derivative_func = sym.lambdify(x_sym, derivative_sym)

# Range of x values
points = range(-5, 6)

for point in points:
    x = np.linspace(point - 0.25, point + 0.25, 1000)
    tangent_line = derivative_func(point) * (x - point) + func_sym.subs(x_sym, point)
    plt.plot(x, tangent_line, color='blue', linewidth=2)
    plt.scatter(point, func_sym.subs(x_sym, point), color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangent Lines and Scatter Points')
plt.grid(True)
plt.show()

for point in points:
    x = np.linspace(point - 0.25, point + 0.25, 1000)
    tangent_line = derivative_func(point) * (x - point) + func_sym.subs(x_sym, point)
    plt.plot(x, tangent_line, color='blue', linewidth=2)
    plt.scatter(point, func_sym.subs(x_sym, point), color='red')
x = np.linspace(-5, 5, 1000)
plt.plot(x, [func(x) for x in x], color="red", linestyle="--")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangent Lines and Function')
plt.grid(True)
plt.show()
```
