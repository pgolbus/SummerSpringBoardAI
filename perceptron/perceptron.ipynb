```python
import numpy as np
import matplotlib.pyplot as plt

from perceptron import Perceptron

# I'mma be honest. I'm not even going to pretend that I read this function. I don't know what it does either
def plot_decision_boundary(perceptron, inputs, labels, title):
    # Plot the input points
    plt.scatter(inputs[:, 0], inputs[:, 1], c=labels, cmap='bwr')

    # Create a meshgrid to plot the decision boundary
    x_min, x_max = inputs[:, 0].min() - 0.1, inputs[:, 0].max() + 0.1
    y_min, y_max = inputs[:, 1].min() - 0.1, inputs[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    grid_inputs = np.c_[xx.ravel(), yy.ravel()]

    # Predict the classes for the grid points
    grid_predictions = np.array([perceptron.predict(x) for x in grid_inputs])
    grid_predictions = grid_predictions.reshape(xx.shape)

    # Plot the decision boundary
    plt.contourf(xx, yy, grid_predictions, cmap='bwr', alpha=0.3)

    # Set plot labels and title
    plt.xlabel('Input 1')
    plt.ylabel('Input 2')
    plt.title(title)

    # Show the plot
    plt.show()

# Training data for AND gate
and_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
and_labels = np.array([0, 0, 0, 1])

# Training data for XOR gate
xor_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
xor_labels = np.array([0, 1, 1, 0])

# Create perceptron objects
and_perceptron = Perceptron(input_size=2)
xor_perceptron = Perceptron(input_size=2)

# Train the perceptrons
and_perceptron.train(and_inputs, and_labels, epochs=10)
xor_perceptron.train(xor_inputs, xor_labels, epochs=10)

# Plot decision boundaries for AND gate
plot_decision_boundary(and_perceptron, and_inputs, and_labels, title="AND Gate")

# Plot decision boundaries for XOR gate
plot_decision_boundary(xor_perceptron, xor_inputs, xor_labels, title="XOR Gate")
```
