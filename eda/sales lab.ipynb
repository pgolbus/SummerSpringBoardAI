```python
# This example uses the np.polyfit function to draw a line through some points

# Generate 2D dataset
np.random.seed(42)

# Number of data points
num_points = 100

# Slope and intercept of the line
true_slope = 2
true_intercept = 5

# Generate x values
x = np.linspace(0, 10, num_points)

# Generate y values with noise
y = true_slope * x + true_intercept + np.random.normal(0, 2, num_points)

# Fit a line to the data
fit_coeffs = np.polyfit(x, y, 1)
fit_slope = fit_coeffs[0]
fit_intercept = fit_coeffs[1]

# Generate y values for the line fit
fit_line = fit_slope * x + fit_intercept

# Plot the dataset and the line fit
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, fit_line, color='red', label='Line Fit')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Dataset with Line Fit')
plt.legend()
plt.grid(True)
plt.show()

print(f"True Line: y = {true_slope} * x + {true_intercept}")
print(f"Line Fit: y = {fit_slope} * x + {fit_intercept}")
```
