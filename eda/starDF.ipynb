```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import quad
```

Probability density function: $P(a \leq X \leq b) = \int\limits_a^b pdf(x)dx$
Cumulative density function: $P(X < k) = \int\limits_{-\infty}^k pdf(t) dt = cdf(k)$

Probability mass function: $P(X = x) = pmf(x)$

Cumulative distribution(?) function: $P(X \leq k) = \sum\limits_{x = -\infty}^k pmf(x) = cdf(k)$



\]

```python
def plot_normal_distribution(mean, sigma):
    x = np.linspace(mean - 4 * sigma, mean + 4 * sigma, 1000)  # Generate x values

    # Calculate PDF and CDF values
    pdf = norm.pdf(x, mean, sigma)
    cdf = norm.cdf(x, mean, sigma)

    # Calculate the integral of PDF and CDF
    pdf_integral, _ = quad(norm.pdf, mean - 4 * sigma, mean + 4 * sigma, args=(mean, sigma))
    cdf_integral, _ = quad(norm.cdf, mean - 4 * sigma, mean + 4 * sigma, args=(mean, sigma))

    # Plotting PDF and CDF on a single figure
    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, 'b-', lw=2, label=f'PDF (Integral: {pdf_integral:.2f})')
    plt.plot(x, cdf, 'r-', lw=2, label=f'CDF (Integral: {cdf_integral:.2f})')
    plt.fill_between(x, pdf, alpha=0.3, color='blue')
    plt.xlabel('x')
    plt.ylabel('Probability / Cumulative Probability')
    plt.title('Normal Distribution')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

def plot_geometric_distribution(p, max_k=10):
    k_values = np.arange(1, max_k+1)  # Values of k for PMF and CDF

    # Calculate PMF
    pmf_values = [(1 - p)**(k - 1) * p for k in k_values]
    pmf_sum = sum(pmf_values)
    
    # Calculate CDF
    cdf_values = [1 - (1 - p)**k for k in k_values]
    cdf_sum = sum(cdf_values)
    
    # Plotting PMF and CDF on a single figure
    plt.figure(figsize=(10, 6))

    plt.bar(k_values, pmf_values, width=.1, color='blue', alpha=0.3, label=f'PMF (Sum: {pmf_sum:0.2f})')
    plt.scatter(k_values, cdf_values, color="red", label=f'CDF (Sum: {cdf_sum:0.2f})')
    plt.xlabel('k')
    plt.ylabel('Probability / Cumulative Probability')
    plt.title('Geometric Distribution')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# Example usag
mean = 0  # Mean of the normal distribution
sigma = .25  # Standard deviation of the normal distribution
plot_normal_distribution(mean, sigma)
```

```python
p = .5
plot_geometric_distribution(p)
```
