```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, kendalltau
```

```python
# Generate correlated data
np.random.seed(42)
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]
X = np.random.multivariate_normal(mean, cov, 1000)
Y = X[:, 0] + 2 * X[:, 1] + np.random.normal(0, 1, 1000)

# Calculate RMSE
rmse = np.sqrt(np.mean((Y - (X[:, 0] + 2 * X[:, 1])) ** 2))

# Calculate correlation coefficients
pearson_rho, _ = pearsonr(X[:, 0], X[:, 1])
spearman_rho, _ = spearmanr(X[:, 0], X[:, 1])
kendall_tau, _ = kendalltau(X[:, 0], X[:, 1])

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='coolwarm')
plt.title(f"Correlated Dataset\nRMSE: {rmse:.2f}, Pearson rho: {pearson_rho:.2f}, Spearman rho: {spearman_rho:.2f}, Kendall tau: {kendall_tau:.2f}")
plt.xlabel("X1")
plt.ylabel("X2")
plt.colorbar(label="Y")
plt.show()
```

```python
# Generate linearly correlated data
np.random.seed(42)
X = np.linspace(0, 10, 1000)
noise = np.random.normal(0, 1, 1000)
Y = 2 * X + 3 + noise

# Calculate RMSE
rmse = np.sqrt(np.mean((Y - (2 * X + 3)) ** 2))

# Calculate correlation coefficients
pearson_rho, _ = pearsonr(X, Y)
spearman_rho, _ = spearmanr(X, Y)
kendall_tau, _ = kendalltau(X, Y)

# Scatter plot
plt.scatter(X, Y, c=Y, cmap='coolwarm')
plt.title(f"Linearly Correlated Dataset\nRMSE: {rmse:.2f}, Pearson rho: {pearson_rho:.2f}, Spearman rho: {spearman_rho:.2f}, Kendall tau: {kendall_tau:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="Y")
plt.show()
```

```python
np.random.seed(42)
X = np.linspace(0, 10, 1000)
Y = np.sin(X) + np.random.normal(0, 0.2, 1000)

# Calculate RMSE
rmse = np.sqrt(np.mean((Y - np.sin(X)) ** 2))

# Calculate correlation coefficients
pearson_rho, _ = pearsonr(X, Y)
spearman_rho, _ = spearmanr(X, Y)
kendall_tau, _ = kendalltau(X, Y)

# Scatter plot
plt.scatter(X, Y, c=Y, cmap='coolwarm')
plt.title(f"Monotonically Correlated Dataset\nRMSE: {rmse:.2f}, Pearson rho: {pearson_rho:.2f}, Spearman rho: {spearman_rho:.2f}, Kendall tau: {kendall_tau:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar(label="Y")
plt.show()
```

```python
# Generate rank correlated data with changing relationship and noise
np.random.seed(42)

# Parameters
n_samples = 1000
n_transition = int(n_samples * 0.95)  # Transition point

# Generate X
X = np.linspace(0, 10, n_samples)

# Generate Y
Y = np.concatenate([
    X[:n_transition] + np.random.normal(0, 0.3, n_transition),  # First part with noise
    np.linspace(10, 7, n_samples - n_transition) + np.random.normal(0, 0.3, n_samples - n_transition)  # Second part with noise
])

# Calculate RMSE
rmse = np.sqrt(np.mean((Y - X) ** 2))

# Calculate correlation coefficients
pearson_rho, _ = pearsonr(X, Y)
spearman_rho, _ = spearmanr(X, Y)
kendall_tau, _ = kendalltau(X, Y)

# Scatter plot
plt.scatter(X, Y, cmap='coolwarm')
plt.title(f"Rank Correlated Dataset (Continuous Hockey Stick with Noise)\nRMSE: {rmse:.2f}, Pearson rho: {pearson_rho:.2f}, Spearman rho: {spearman_rho:.2f}, Kendall tau: {kendall_tau:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
