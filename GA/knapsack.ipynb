```python
from matplotlib import pyplot as plt

from knapsack import KnapsackGA
```

```python
# Genetic Algorithm parameters
POPULATION_SIZE = 10
MUTATION_RATE = 0.01
GENERATIONS = 1000

# Knapsack parameters
MAX_WEIGHT = 15
# List of items [weight, value]
KNAPSACK = [
    [1, 5],
    [2, 6],
    [2, 10],
    [3, 12],
    [3, 14],
    [4, 15],
    [5, 16],
    [5, 11],
    [6, 17],
]

# plotting parameters
PLOT = 5
```

```python
ga_runner = KnapsackGA(KNAPSACK)
for i in range(PLOT):
    scores = ga_runner.genetic_algorithm(GENERATIONS, POPULATION_SIZE, MUTATION_RATE, max_weight=MAX_WEIGHT)
    plt.plot(scores)
    max_score = max(scores)
    plt.ylim(0, max_score+4)
    plt.title(f"\nMax score: {max_score}")
    plt.show()
    if i != PLOT - 1:
        print("\n\n-----------------------------------------------------------------------------------")
```
