```python
from matplotlib import pyplot as plt

from knapsack import KnapsackGA
```

```python
# Genetic Algorithm parameters
POPULATION_SIZE = 100
MUTATION_RATE = 0.01
GENERATIONS = 1000

# Knapsack parameters
MAX_WEIGHT = 15
# List of items [weight, value]
KNAPSACK = [
    [2, 6],
    [2, 10],
    [3, 12],
    [4, 15],
    [5, 16]
]

# plotting parameters
PLOT = 5
```

```python
ga_runner = KnapsackGA(KNAPSACK)
for i in range(PLOT):
    scores = ga_runner.genetic_algorithm(GENERATIONS, POPULATION_SIZE, MUTATION_RATE, max_weight=MAX_WEIGHT)
    plt.scatter(range(len(scores)), scores)
    max_score = max(scores)
    plt.title(f"\nMax score: {max_score}")
    plt.show()
    if i != PLOT - 1:
        print("\n\n-----------------------------------------------------------------------------------")
```
