```python
from matplotlib import pyplot as plt
import numpy as np

from knapsack import generate_knapsack, KnapsackGA
```

```python
# Genetic Algorithm parameters
POPULATION_SIZE = 10
MUTATION_RATE = 0.01
CROSSOVER_RATE = 0.7
GENERATIONS = 25

# Knapsack parameters
MAX_WEIGHT = 500
# List of items [weight, value]
KNAPSACK = generate_knapsack(100)

# plotting parameters
PLOT = 5
```

```python
ga_runner = KnapsackGA(KNAPSACK)
max_scores = []
for i in range(PLOT):
    scores = ga_runner.genetic_algorithm(GENERATIONS, POPULATION_SIZE, MUTATION_RATE, CROSSOVER_RATE, max_weight=MAX_WEIGHT)
    plt.plot(scores)
    max_score = max(scores)
    max_scores.append(max_score)
    plt.ylim(min(scores)-100, max_score+100)
    plt.title(f"\nMax score: {max_score}")
    plt.show()
    if i != PLOT - 1:
        print("\n\n-----------------------------------------------------------------------------------")
np.mean(max_scores)
```
