```python
# Load the data from the CSV file
df = pd.read_csv('weather_data.csv')

# Print the first 5 rows of the dataframe
df.head()
```

```python
# Get a summary of the dataframe
df.info()
df.describe()
```

## Part 2: Data Manipulation with Pandas

1. Use the `value_counts` method to find the most common wind direction.
2. Use the `groupby` method to find the average temperature for each month.
3. Use the `loc` method to select data for a specific date.
4. Use the `iloc` method to select the first 10 rows of the dataframe.
5. Use the `apply` method to apply a function to the temperature columns that converts temperatures from Celsius to Fahrenheit.


```python
# 1. Find the most common wind direction
most_common_wind_direction = df['Data.Wind.Direction'].value_counts().idxmax()
print('Most Common Wind Direction:', most_common_wind_direction)

# 2. Find the average temperature for each month
avg_temp_per_month = df.groupby('Date.Month')['Data.Temperature.Avg Temp'].mean()
print('\nAverage Temperature per Month:\n', avg_temp_per_month)

# 3. Select data for a specific date
specific_date_data = df.loc[df['Date.Full'] == '2022-01-01']
print('\nData for 2022-01-01:\n', specific_date_data)

# 4. Select the first 10 rows of the dataframe
first_10_rows = df.iloc[:10]
print('\nFirst 10 Rows:\n', first_10_rows)

# 5. Convert temperatures from Celsius to Fahrenheit
def celsius_to_fahrenheit(temp):
    return temp * 9/5 + 32

df['Data.Temperature.Avg Temp'] = df['Data.Temperature.Avg Temp'].apply(celsius_to_fahrenheit)
df['Data.Temperature.Max Temp'] = df['Data.Temperature.Max Temp'].apply(celsius_to_fahrenheit)
df['Data.Temperature.Min Temp'] = df['Data.Temperature.Min Temp'].apply(celsius_to_fahrenheit)

df.head()
```

## Part 3: Data Visualization with Pandas

1. Use the `plot` method to create a line plot of the average temperature over time.
2. Use the `hist` method to create a histogram of wind speeds.
3. Use the `scatter` method to create a scatter plot of average temperature vs. precipitation.
4. Use the `boxplot` method to create a box plot of temperatures for each month.


```python
# Fit a second-degree polynomial to the average temperature over time
poly_coeffs = np.polyfit(df.index, df['Data.Temperature.Avg Temp'], deg=2)
poly_curve = np.poly1d(poly_coeffs)

# Scatter plot of average temperature over time
df['Data.Temperature.Avg Temp'].plot(kind='line')
plt.title('Average Temperature Over Time')
plt.xlabel('Days')
plt.ylabel('Temperature (Fahrenheit)')

# Add polynomial curve
plt.plot(df.index, poly_curve(df.index), color='red', label='Second-degree Polynomial Fit')

# Add legend
plt.legend()
plt.show()
a,b,c = poly_coeffs
print(f"temp = {a:.6f} days^2 + {b:.6f} days + {c:.6f}")
```

### We have discovered seasons!

Note that this is a linear combination of variables and coefficients. This quadratic curve is _still a linear model!_

```python
# 2. Histogram of wind speeds
df['Data.Wind.Speed'].hist(bins=10)
plt.title('Histogram of Wind Speeds')
plt.xlabel('Wind Speed')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Fit a linear regression model to the data
X = df['Data.Temperature.Avg Temp'].values.reshape(-1, 1)  # Reshape because this is a single feature
y = df['Data.Precipitation'].values

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)

# Scatter plot of average temperature vs. precipitation
df.plot(kind='scatter', x='Data.Temperature.Avg Temp', y='Data.Precipitation')
plt.title('Average Temperature vs. Precipitation')
plt.xlabel('Temperature (Fahrenheit)')
plt.ylabel('Precipitation')

# Add regression line
plt.plot(X, y_pred, color='red', label=f'Regression line (R^2 = {r2:.2f})')

# Add legend
plt.legend()
plt.show()
```

```python
# 4. Box plot of temperatures for each month
df.boxplot(column=['Data.Temperature.Avg Temp'], by='Date.Month')
plt.title('Box Plot of Temperatures for Each Month')
plt.xlabel('Month')
plt.ylabel('Temperature (Fahrenheit)')
plt.suptitle('')  # Suppress the automatic 'Boxplot grouped by' title
plt.show()
```
